# Linear Models

In this notebook you'll get acquainted with linear models and with data modeling pipeline. Further reading is available at [sklearn documentation](https://scikit-learn.org/stable/modules/linear_model.html).

## Prerequisites

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
from sklearn import datasets

sns.set(style="darkgrid")

%matplotlib inline
```

## Load Data

Load the boston house-prices dataset. We'll use it to demonstrate how regression works. Attribute information can be found by link: https://scikit-learn.org/stable/datasets/index.html#boston-dataset

```python
# Load dataset
data = datasets.load_boston()
X, y = pd.DataFrame(data.data, columns=data.feature_names), data.target

data = X.copy()
data["target"]=y
column_names = X.columns
```

## Model Training

He we'll train a simple linear regression model. This stage usually contains following steps:

1. Split dataset into train/test sets. Usual ratio is about 70%/30% or 80%/20%.
2. Preprocess data and extract features. Here we'll just standardize features, using [StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html).
3. Train model, using train set.
4. Compute model error on test set.

```python
# Lets train linear regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# Scale Data
scaler = StandardScaler()
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)
X_train, X_test = scaler.fit_transform(X_train), scaler.fit_transform(X_test)

lr = LinearRegression()
lr.fit(X_train, y_train);

# Compute Test Error
test_mse = mean_squared_error(lr.predict(X_test), y_test)
print("MSE on test : {}".format(test_mse))
```

```python
# Log target 
lr = LinearRegression()
lr.fit(X_train, np.log(y_train));
test_mse = mean_squared_error(np.exp(lr.predict(X_test)), y_test)
print("MSE on test : {}".format(test_mse))
```

```python
# Visualize model weights
plt.figure(figsize=(10, 10))
coeffs = lr.coef_
sns.barplot(x=column_names, y=coeffs);
```

## SVM

Here we'll try to model data with Support Vector Machines. Additional information can be found at [sklearn documentation](https://scikit-learn.org/stable/modules/svm.html).

```python
# Let's generate some synthetic data for classfication task
first_class_X = np.random.normal(size=(100,2), loc=0)
second_class_X = np.random.normal(size=(100,2), loc=4)
```

```python
plt.figure(figsize=(8,8))
plt.plot(first_class_X[:,0], first_class_X[:,1], ".");
plt.plot(second_class_X[:,0], second_class_X[:,1], ".");
```

You can see that dataset - linear separable

```python
# Create labels 
f_class_y = np.zeros(100)
s_class_y = np.ones(100)
```

```python
# Concat data
X = np.concatenate([first_class_X, second_class_X])
y = np.concatenate([f_class_y, s_class_y])
```

```python
# Split on train/test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc = SVC(kernel="linear", C=0.5)
svc.fit(X_train, y_train)
```

```python
print("Accuracy on test data : {}".format(accuracy_score(y_test, svc.predict(X_test))))
```

```python
# Visualize support vectors
vectors = svc.support_vectors_

plt.figure(figsize=(8,8))
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap="winter");
plt.scatter(vectors[:,0], vectors[:,1], marker="*",s=10**3, c="red");
```
